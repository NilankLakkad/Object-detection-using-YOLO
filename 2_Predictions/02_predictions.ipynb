## Real Time Object Detection

```python
cap = cv2.VideoCapture('video.mp4')


while True:
    ret, frame = cap.read()
    if ret == False:
        print('unable to read video')
        break
        
    pred_image = yolo.predictions(frame)
    
    cv2.imshow('YOLO',pred_image)
    if cv2.waitKey(1) == 27:
        break
        
cv2.destroyAllWindows()
cap.release()
```

```python
import cv2

# Initialize webcam video capture (use 0 for the default camera)
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        print("Unable to read from camera")
        break
    
    # Perform YOLO predictions on the frame
    pred_image = yolo.predictions(frame)  # Replace with your YOLO prediction function

    # Display the frame with predictions
    cv2.imshow('YOLO Real-Time Object Detection', pred_image)

    # Break the loop if 'Esc' key is pressed
    if cv2.waitKey(1) == 27:
        break

# Release resources
cv2.destroyAllWindows()
cap.release()
```
